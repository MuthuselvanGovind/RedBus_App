```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import time

# Initialize lists to store route names and links
Route_Name = []
Route_Link = []

# Initialize the WebDriver
driver = webdriver.Chrome()

# Open the URL
driver.get('https://www.redbus.in/online-booking/rsrtc/?utm_source=rtchometile')
driver.maximize_window()

# Wait until the elements are present
wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, "route")))

# Scroll down to ensure elements are in view
driver.execute_script("window.scrollTo(0, 500);")
time.sleep(3)  # You can use a wait here if necessary

def getText():
    elements = driver.find_elements(By.CLASS_NAME, "route")
    for element in elements:
        Route_Name.append(element.text)
        Route_Link.append(element.get_attribute('href'))
getText()
# Pagination handling
for page_num in range(1, 4):
    try:
        pagination_container = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="root"]/div/div[4]/div[12]')))
        next_page_button = pagination_container.find_element(By.XPATH,f'.//div[contains(@class,"DC_117_pageTabs") and text()="{page_num + 1}"]')

        actions = ActionChains(driver)
        actions.move_to_element(next_page_button).perform()
        time.sleep(2)
        driver.execute_script("arguments[0].scrollIntoView();", next_page_button)
        time.sleep(2)
        next_page_button.click()
        print("Success")
        getText()
    except Exception as e:
        print(f"An error occurred: {e}")

print("Route Names are:\n", "\n".join(Route_Name))
print("Route Links are:\n", "\n".join(Route_Link))


# Close the WebDriver
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Initialize lists to store bus information
Route_Name_1 = []
Route_Link_1 = []
Bus_names = []
Bus_types = []
Start_times = []
End_times = []
Total_durations = []
Ratings = []
Prices = []
Seat_availabilities = []

# Sample Route_Name list for demonstration (replace with your actual list)
RouteName = [
    'Burdwan to Kolkata',
 'Kolkata to Burdwan',
 'Durgapur (West Bengal) to Kolkata',
 'Kolkata to Haldia',
 'Haldia to Kolkata',
 'Kolkata to Durgapur (West Bengal)',
 'Kolkata to Arambagh (West Bengal)',
 'Digha to Kolkata',
 'Kolkata to Digha',
 'Kolkata to Bankura',
 'Asansol (West Bengal) to Kolkata',
 'Midnapore to Kolkata',
 'Kolkata to Asansol (West Bengal)',
 'Kolkata to Siliguri',
 'Kolkata to Nimtouri',
 'Siliguri to Kolkata',
 'Kolkata to Contai (Kanthi)',
 'Digha to Durgapur (West Bengal)',
 'Kolkata to Midnapore',
 'Kolkata to Nandakumar (west bengal)',
 'Kolkata to Mecheda (West Bengal)',
 'Durgapur (West Bengal) to Digha',
 'Kolkata to Kolaghat',
 'Jhargram to Kolkata',
 'Digha to Barasat (West Bengal)',
 'Durgapur (West Bengal) to Siliguri',
 'Midnapore to Barasat (West Bengal)',
 'Kolkata to Heria',
 'Kolkata to Chandipur (West Bengal)',
 'Siliguri to Durgapur (West Bengal)',
 'Berhampore (West Bengal) to Durgapur (West Bengal)',
 'Berhampore (West Bengal) to Suri',
 'Barasat (West Bengal) to Digha',
 'Kolkata to Futishanko',
 'Barasat (West Bengal) to Midnapore',
 'Durgapur (West Bengal) to Berhampore (West Bengal)',
 'Durgapur (West Bengal) to Malda',
 'Digha to Asansol (West Bengal)',
 'Kolkata to Ramnagar (West Bengal)',
 'Kolkata to Bajkul (West Bengal)',
 'Kolkata to Kirnahar (West Bengal)',
 'Durgapur (West Bengal) to Bankura',
 'Siliguri to Malda',
 'Kolkata to Suri',
 'Purulia to Durgapur (West Bengal)'
]



# Ensure Route_Name and Route_Link have the same length
if len(RouteName) != len(Route_Link):
    raise ValueError("Route_Name and Route_Link lists must have the same length.")

# Initialize the WebDriver
driver = webdriver.Chrome()
driver.maximize_window()

# Validate URLs
def is_valid_url(url):
    return isinstance(url, str) and url.startswith("http")

# Getting link
for index, Link in enumerate(Route_Link):
    if not is_valid_url(Link):
        print(f"Invalid URL: {Link}")
        continue  # Skip invalid URL
    
    driver.get(Link)
    time.sleep(2)
    
    # Click elements on bus
    click_elements = driver.find_elements(By.XPATH, f"//a[contains(@href, '{Link}')]")
    for click_element in click_elements:
        driver.execute_script("arguments[0].scrollIntoView(true);", click_element)
        time.sleep(1)
        try:
            click_element.click()
        except Exception as e:
            print(f"Exception while clicking: {e}")
            driver.execute_script("arguments[0].click();", click_element)
        time.sleep(2)
    
    view_bus = driver.find_elements(By.XPATH, "//div[@class='button']")
    for element_click in view_bus:
        driver.execute_script("arguments[0].scrollIntoView(true);", element_click)
        time.sleep(1)
        try:
            element_click.click()
        except Exception as e:
            print(f"Exception while clicking: {e}")
            driver.execute_script("arguments[0].click();", element_click)
        time.sleep(2)
    
    scrolling = True
    while scrolling:
        old_page_source = driver.page_source
        print("Old page source captured")
        
        body = driver.find_element(By.TAG_NAME, "body")
        body.send_keys(Keys.PAGE_DOWN)
        time.sleep(5)                            
        new_page_source = driver.page_source
        if new_page_source == old_page_source:
            scrolling = False

    # Print the current page source for debugging
    print(driver.page_source)
    
    wait = WebDriverWait(driver, 30)  # Increased the timeout to 30 seconds
    try:
        wait.until(EC.presence_of_all_elements_located((By.XPATH, "//div[@class='travels lh-24 f-bold d-color']")))
        wait.until(EC.presence_of_all_elements_located((By.XPATH, "//div[@class='bus-type f-12 m-top-16 l-color evBus']")))
        wait.until(EC.presence_of_all_elements_located((By.XPATH, "//*[@class='dp-time f-19 d-color f-bold']")))
        wait.until(EC.presence_of_all_elements_located((By.XPATH, "//*[@class='bp-time f-19 d-color disp-Inline']")))
        wait.until(EC.presence_of_all_elements_located((By.XPATH, "//*[@class='dur l-color lh-24']")))
        wait.until(EC.presence_of_all_elements_located((By.XPATH, '//div[@class="rating-sec lh-24"]')))
        wait.until(EC.presence_of_all_elements_located((By.XPATH, "//*[@class='fare d-block']")))
        wait.until(EC.presence_of_all_elements_located((By.XPATH, "//*[@class='seat-left m-top-30']")))
        
        # Extract bus details
        Bus_name = driver.find_elements(By.XPATH, "//div[@class='travels lh-24 f-bold d-color']")
        Bus_type = driver.find_elements(By.XPATH, "//div[@class='bus-type f-12 m-top-16 l-color evBus']")
        Start_time = driver.find_elements(By.XPATH, "//*[@class='dp-time f-19 d-color f-bold']")
        End_time = driver.find_elements(By.XPATH, "//*[@class='bp-time f-19 d-color disp-Inline']")
        Total_duration = driver.find_elements(By.XPATH, "//*[@class='dur l-color lh-24']")
        Rating = driver.find_elements(By.XPATH, '//div[@class="rating-sec lh-24"]')
        Price = driver.find_elements(By.XPATH, "//*[@class='fare d-block']")
        Seats = driver.find_elements(By.XPATH, "//*[@class='seat-left m-top-30']")
    
        for Bus in Bus_name:
            Bus_names.append(Bus.text)
            Route_Name_1.append(RouteName[index])  # Append the corresponding route name
            Route_Link_1.append(Link)               # Append the link
        for Btype in Bus_type:
            Bus_types.append(Btype.text)
        for Stime in Start_time:
            Start_times.append(Stime.text)
        for Etime in End_time:
            End_times.append(Etime.text)
        for Duration in Total_duration:
            Total_durations.append(Duration.text)
        for Rate in Rating:
            Ratings.append(Rate.text)                                    
        for Ps in Price:
            Prices.append(Ps.text)
        for Sa in Seats:
            Seat_availabilities.append(Sa.text)
    except Exception as e:
        print(f"An error occurred while waiting for elements: {e}")

# Ensure all elements in lists are strings
Route_Name_1 = [str(item) for item in Route_Name_1]
Route_Link_1 = [str(item) for item in Route_Link_1]
Bus_names = [str(item) for item in Bus_names]
Bus_types = [str(item) for item in Bus_types]
Start_times = [str(item) for item in Start_times]
End_times = [str(item) for item in End_times]
Total_durations = [str(item) for item in Total_durations]
Ratings = [str(item) for item in Ratings]
Prices = [str(item) for item in Prices]
Seat_availabilities = [str(item) for item in Seat_availabilities]

# Print collected data
print("Route Names are:\n", "\n".join(Route_Name_1)) 
print("Route Links are:\n", "\n".join(Route_Link_1))
print("Bus Names are:\n", "\n".join(Bus_names)) 
print("Bus Types are:\n", "\n".join(Bus_types))                                  
print("Start Times are:\n", "\n".join(Start_times))                                   
print("End Times are:\n", "\n".join(End_times))                                   
print("Total Durations are:\n", "\n".join(Total_durations))                                   
print("Ratings are:\n", "\n".join(Ratings))                                   
print("Prices are:\n", "\n".join(Prices))                                   
print("Seat Availabilities are:\n", "\n".join(Seat_availabilities)) 

# Close the WebDriver
driver.quit()
```

```python
import pandas as pd

max_length = max(
    len(Route_Name_1), len(Route_Link_1), len(Bus_names), len(Bus_types), 
    len(Start_times), len(End_times), len(Total_durations), len(Ratings), 
    len(Prices), len(Seat_availabilities)
)

# Function to pad lists to the maximum length
def pad_list(lst, max_length):
    return lst + [None] * (max_length - len(lst))

# Pad each list to the maximum length
Route_Name_1 = pad_list(Route_Name_1, max_length)
Route_Link_1 = pad_list(Route_Link_1, max_length)
Bus_names = pad_list(Bus_names, max_length)
Bus_types = pad_list(Bus_types, max_length)
Start_times = pad_list(Start_times, max_length)
End_times = pad_list(End_times, max_length)
Total_durations = pad_list(Total_durations, max_length)
Ratings = pad_list(Ratings, max_length)
Prices = pad_list(Prices, max_length)
Seat_availabilities = pad_list(Seat_availabilities, max_length)


RAJRTC_Route = {
    'RouteName': Route_Name_1,
    'RouteLink': Route_Link_1,
    'Bus Name': Bus_names,
    'Bus Type': Bus_types,
    'Start Time': Start_times,
    'End Time': End_times,
    'Total Duration': Total_durations,
    'Rating': Ratings,
    'Price': Prices,
    'Seat Availability': Seat_availabilities
}

RAJRTC_Table = pd.DataFrame(RAJRTC_Route)
```
